```python
#Import dependancies
import pandas as pd
import numpy as np

#Sets csv file location
file = "Resources/purchase_data.csv"

#Converts the csv file into a dataframe
purchase_data = pd.read_csv(file)

#Displays the csv dataframe to confirm it's been read properly
purchase_data.head()
```

```python
#Removes rows that contain duplicate player SNs
purchase_data_uniquePlayers = purchase_data.drop_duplicates(subset='SN',keep='last')

#Creates a dataframe of every unique player SN along with their associated age
uniquePlayers = purchase_data_uniquePlayers[["SN", "Age"]]

#Creates a number variable from the previously created dataframe of all players
totalPlayers = uniquePlayers["SN"].count()

#Creates the summary dataframe that displays the amount of total players
totalPlayers_summary = pd.DataFrame({"Total Players": [totalPlayers]})


#Displays total amount of players
totalPlayers_summary
```

```python
#Removes rows that contain duplicate item IDs
purchase_data_uniqueItems = purchase_data.drop_duplicates(subset='Item ID',keep='last')

#Creates a new Dataframe with all unique Item IDs
uniqueItems = purchase_data_uniqueItems[["Item ID"]]

#Sets a variable for the total number of unique items
totalItems = uniqueItems["Item ID"].count()

#Gets average Item price and sets a variable
averagePrice = round(purchase_data["Price"].mean(), 2)

#Gets total number of purchases and sets a variable
totalPurchases = purchase_data["Purchase ID"].count()

#Sums total revenue and sets a variable
totalRevenue = purchase_data["Price"].sum()

#Creates a summary table using the variables totalItems, averagePrice, totalPurchases, and totalRevenue
summary_table = pd.DataFrame({"Unique Items": [totalItems],
                              "Average Price": [averagePrice],
                             "Total Purchases": [totalPurchases],
                             "Total Revenue": [totalRevenue]})

#Maps the values in summary_table to have $ values
summary_table["Average Price"] = summary_table["Average Price"].map("${:.2f}".format)
summary_table["Total Revenue"] = summary_table["Total Revenue"].map("${:.2f}".format)


#Displays a table of item values
summary_table
```

```python
#Creates dataframes for all male players, female players, and other players
malePlayers = purchase_data_uniquePlayers.loc[purchase_data_uniquePlayers["Gender"] == "Male"]
femalePlayers = purchase_data_uniquePlayers.loc[purchase_data_uniquePlayers["Gender"] == "Female"]
otherPlayers = purchase_data_uniquePlayers.loc[purchase_data_uniquePlayers["Gender"] == "Other / Non-Disclosed"]

#Counts the number of male, female, and other players using their unique dataframes
maleCount = malePlayers["Gender"].count()
femaleCount = femalePlayers["Gender"].count()
otherCount = otherPlayers["Gender"].count()

#Calculates percentages of genders
malePercent = round((maleCount / totalPlayers) * 100, 2)
femalePercent = round((femaleCount / totalPlayers) * 100, 2)
otherPercent = round((otherCount / totalPlayers) * 100, 2)

#Creates string variables for genders to use for summary
Male = "Male"
Female = "Female"
Other = "Other"

#Creates the summary table for displaying gender counts and percentages
gender_summary = pd.DataFrame({"Gender": [Male] + [Female] + [Other],
                              "Counts": [maleCount] + [femaleCount] + [otherCount],
                              "Percentages": [malePercent] + [femalePercent] + [otherPercent]})

#Displays counts and percentages for genders
gender_summary
```

```python
#Creates variables for male statistics
malePurchases = malePlayers["Purchase ID"].count()
maleAveragePrice = malePlayers["Price"].mean()

#Creates variables for female statistics
femalePurchases = femalePlayers["Purchase ID"].count()
femaleAveragePrice = femalePlayers["Price"].mean()

#Creates variables for other statistics
otherPurchases = otherPlayers["Purchase ID"].count()
otherAveragePrice = otherPlayers["Price"].mean()

#Creates the summary dataframe
purchase_analysis = pd.DataFrame({"Gender": [Male] + [Female] + [Other],
                                 "Total Purchases": [malePurchases] + [femalePurchases] + [otherPurchases],
                                 "Average price": [maleAveragePrice] + [femaleAveragePrice] + [otherAveragePrice]})

#Maps the summary dataframe to include $ values
purchase_analysis["Average price"] = purchase_analysis["Average price"].map("${:.2f}".format)

#Displays statistics relating to purchases based on gender
purchase_analysis
```

```python
#Creates a groupby of all player SNs
groupedSN = purchase_data.groupby("SN")

#Creates a dataframe of the sum of player SNs (how many purchases each SN made)
top_spenders = groupedSN.sum()

#Sorts the dataframe so the highest purchase price comes first
top_spenders = top_spenders.sort_values("Price", ascending=False)

#Renames price to total purchase value
top_spenders = top_spenders.rename(columns={"Price": "Total Purchase Value"})

#Drops unneccessary columns included from the groupby
top_spenders = top_spenders.drop(["Purchase ID", "Age", "Item ID"], axis=1)

#Creates a column that lists how many purchases were made based on how many times each SN appeared
top_spenders["Purchases"] = purchase_data["SN"].value_counts()

#Finds the average purchase price per player by dividing their total spent by their amount of purchases
top_spenders["Average Purchase Price"] = round(top_spenders["Total Purchase Value"] / top_spenders["Purchases"], 2)

#Maps values to have $ symbols
top_spenders["Total Purchase Value"] = top_spenders["Total Purchase Value"].map("${:.2f}".format)
top_spenders["Average Purchase Price"] = top_spenders["Average Purchase Price"].map("${:.2f}".format)

#Displays the summary table
top_spenders.head()
```

```python
#Creates a groupby of all item names
groupedItems = purchase_data.groupby("Item Name")

#Creates a dataframe of all item names
popular_items = groupedItems.sum()

#Creates a column for purchase count by showing how many times the item appeared in the initial dataframe
popular_items["Purchase Count"] = purchase_data["Item Name"].value_counts()

#Sorts the purchase count column to have the most purchases show up first
popular_items = popular_items.sort_values("Purchase Count", ascending=False)

#Drops unnecessary columns included with the groupby
popular_items = popular_items.drop(["Purchase ID", "Age", "Item ID"], axis=1)

#Renames the Price column to Total Purchase Value
popular_items = popular_items.rename(columns={"Price": "Total Purchase Value"})

#Creates a column for item price by dividing the total purchase value by the number of purchases
popular_items["Item Price"] = round(popular_items["Total Purchase Value"] / popular_items["Purchase Count"], 2)

#Reorganizes the column order
popular_items = popular_items[["Purchase Count", "Item Price", "Total Purchase Value"]]

#Maps monetary values to have a $ symbol
popular_items["Item Price"] = popular_items["Item Price"].map("${:.2f}".format)
popular_items["Total Purchase Value"] = popular_items["Total Purchase Value"].map("${:.2f}".format)

#Displays the summary dataframe
popular_items.head()
```

```python
#Creates a profitable items dataframe
profitable_items = groupedItems.sum()

#Drops unnecessary columns from groupedItems
profitable_items = profitable_items.drop(["Purchase ID", "Age", "Item ID"], axis=1)

#Merges in columns from the popular items dataframe
profitable_items = pd.merge(profitable_items, popular_items, on="Item Name")

#Drops the duplicate Total Purchase Value column
profitable_items = profitable_items.drop(["Total Purchase Value"], axis=1)

#Sorts by price
profitable_items = profitable_items.sort_values("Price", ascending=False)

#Renames price to Total Purchase Value (couldn't sort the old one since it had string values due to mapping)
profitable_items = profitable_items.rename(columns={"Price": "Total Purchase Value"})

#Reorganizes the columns
profitable_items = profitable_items[["Purchase Count", "Item Price", "Total Purchase Value"]]

#Maps the newly sorted Total Purchase Value column to include a $ symbol
profitable_items["Total Purchase Value"] = profitable_items["Total Purchase Value"].map("${:.2f}".format)

#Displays the summary table
profitable_items.head()
```
